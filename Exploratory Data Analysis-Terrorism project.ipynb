## RENAMING THE COLUMNS

```python
data.rename(columns={'iyear':'Year',
                     'imonth':'Month',
                     'iday':'Day',
                     'country_txt':'Country',
                     'provstate':'state',
                     'region_txt':'Region',
                     'attacktype1_txt':'AttackType',
                     'target1':'Target',
                     'nkill':'Killed',
                     'nwound':'Wounded',
                     'summary':'Summary',
                     'gname':'Group',
                     'targtype1_txt':'Target_type',
                     'weaptype1_txt':'Weapon_type',
                     'motive':'Motive'},inplace=True)
```

```python
data=data[['Year','Month','Day','Country','state','Region','city','latitude','longitude','AttackType','Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
```

```python
data.head()  # Extract new data
```

```python
data.info()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data["Killed"]=data["Killed"].fillna(0)
data["Wounded"]=data["Wounded"].fillna(0)
data["Casualty"]=data["Killed"]+data["Wounded"]
```

```python
data.describe()
```

# DATA VISUALIZATION

```python
data.hist(figsize=(20,10))
```

## CORRELATION ANALYSIS

```python
plt.figure(figsize=(15,10))
sns.heatmap(np.round(data.corr(),2),annot=True, cmap='coolwarm')
```

## Number of Attacks in each Year

```python
plt.subplots(figsize=(15, 6))
sns.countplot('Year', data=data, palette='Blues_d', edgecolor='k') 
plt.xticks(rotation=70)
plt.title('Number Of Terrorist Activities Each Year')
plt.show()
```

```python
#Killed in each Year
yk=data[["Year","Killed"]].groupby("Year").sum()
```

```python
#Wounded in each Year
xy=data[["Year","Wounded"]].groupby("Year").sum()
```

```python
fig=plt.figure()
ax0=fig.add_subplot(2,1,1)
ax1=fig.add_subplot(2,1,2)
#Killed
yk.plot(kind="bar",color="green",figsize=(15,15),ax=ax0)
ax0.set_title("People Killed in each Year")
ax0.set_xlabel("Years")
ax0.set_ylabel("Number of People Killed") 
#Wounded
xy.plot(kind="bar",color="cornflowerblue",figsize=(15,15),ax=ax1)
ax1.set_title("People Wounded in each Year")
ax1.set_xlabel("Years")
ax1.set_ylabel("Number of People Wounded")

plt.show()
```

## Terrorist Activities by Region in each Year Using Area Plot

```python
pd.crosstab(data.Year, data.Region).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
```

## Country wise Attacks

```python
ct=data["Country"].value_counts().head(10)
ct
```

```python
ct.plot(kind="bar",color="#FF33DD",figsize=(15,6))
plt.title("Country wise Attacks",fontsize=13)
plt.xlabel("Countries",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Attacks",fontsize=13)
plt.show()
```

## Target Type wise Attacks 

```python
plt.figure(figsize=(13,6))
sns.countplot(data['Target_type'],order=data['Target_type'].value_counts().index,
              palette='magma')
plt.xticks(rotation=90)
plt.xlabel('Type')
plt.title('Type of Target')
plt.show()
```

## Top 10 Country with most Attacks

```python
top_10_country=data.Country.value_counts()[:10].to_frame().reset_index()
top_10_country
```

```python
top_10_country.columns=['country_name','count']
```

```python
top_10_country
plt.figure(figsize=(15,5))
sns.barplot(data=top_10_country,x='country_name',y='count',ec='black',lw=1)
plt.grid(True)
plt.title('Top 10 Countries with most Attacks',fontsize=20)
plt.xlabel('Country',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.show()
```

## Methods of Attack

```python
plt.figure(figsize=(15,6))
sns.countplot(data['AttackType'],order=data['AttackType'].value_counts().index,
              palette='magma')
plt.xticks(rotation=60)
plt.xlabel('Method')
plt.title('Method of Attack')
plt.show()
```
